```python
from spacy import displacy
from IPython.display import display, HTML
import spacy

nlp = spacy.load('es_core_news_lg')
def procesar_noticia_y_guardar_entidades(noticia, archivo_salida):
    doc = nlp(noticia)
    entidades = []
    for ent in doc.ents:
        # Extraer los POS de cada token en la entidad
        pos_tags = [token.pos_ for token in ent]
        entidades.append((ent.text, ent.label_, pos_tags))

    # Guardar entidades en un archivo
    with open(archivo_salida, "w", encoding="utf-8") as f:
        for texto, etiqueta, pos_tags in entidades:
            pos_str = ",".join(pos_tags)
            f.write(f"{texto}\t{etiqueta}\t{pos_str}\n")
    
    # Visualizar la nota etiquetada
    print("\nVisualización:")
    html = displacy.render(doc, style="ent", jupyter=False)

    display(HTML(html))
```
